This code loads the images and annotations from the DataFrame created in the previous code block.

It initializes two empty lists images and annotations to store the image data and annotations, respectively.

It loops through each row in the annotations_df DataFrame using the iterrows() method. For each row, it loads the image using the cv2.imread() function and the filename stored in the "filename" column of the DataFrame. It then loads the annotation information from the "xmin", "ymin", "xmax", "ymax", and "class" columns of the DataFrame into a dictionary called annotation. This dictionary is structured to match the expected format for annotations in object detection tasks, with a list of bounding boxes and a list of corresponding labels.

Finally, it appends the image and annotation to the images and annotations lists, respectively.

Overall, this code loads the images and annotations from the fracture1/train directory and _annotations.csv file into images and annotations lists, respectively, using OpenCV to load the images and a pandas DataFrame to load the annotations.

```python
# Load the images and annotations:
images = []
annotations = []

for index, row in annotations_df.iterrows():
    # Load the image
    image = cv2.imread(os.path.join(image_folder, row["filename"]))

    # Load the annotation
    xmin = row["xmin"]
    ymin = row["ymin"]
    xmax = row["xmax"]
    ymax = row["ymax"]
    label = row["class"]
    annotation = {
        "boxes": [[xmin, ymin, xmax, ymax]],
        "labels": [label]
    }

    images.append(image)
    annotations.append(annotation)
```

This code preprocesses the loaded images and annotations.

It initializes two empty lists preprocessed_images and preprocessed_annotations to store the preprocessed image data and annotations, respectively.

It loops through each image and annotation pair in zip(images, annotations). For each pair, it resizes the image to a shape of (224, 224) using the cv2.resize() function. It then converts the annotation data to the format expected by the RCNN model, which is a dictionary with keys "boxes" and "labels", where "boxes" is an array of bounding boxes and "labels" is an array of corresponding labels.

Finally, it appends the preprocessed image and annotation to the preprocessed_images and preprocessed_annotations lists, respectively.

After the loop, it converts the preprocessed data to numpy arrays using the np.array() function.

```python
#Preprocess the images and annotations:
import numpy as np

# Preprocess the images and annotations
preprocessed_images = []
preprocessed_annotations = []

for image, annotation in zip(images, annotations):
    # Resize the image
    image = cv2.resize(image, (224, 224))

    # Convert the annotation to the format expected by the RCNN model
    boxes = np.array(annotation["boxes"])
    labels = np.array(annotation["labels"])
    parsed_annotation = {
        "boxes": boxes,
        "labels": labels
    }

    preprocessed_images.append(image)
    preprocessed_annotations.append(parsed_annotation)

# Convert the preprocessed data to numpy arrays
preprocessed_images = np.array(preprocessed_images)
preprocessed_annotations = np.array(preprocessed_annotations)
```

This code defines a function to load images and their corresponding bounding boxes and labels from a directory and CSV file containing the bounding box coordinates and labels. The function load_data takes as input the directory where the images are located and a Pandas DataFrame containing the bounding box coordinates and labels.

Within the function, it uses the load_img function from Keras to load the image and resizes it to 224x224 pixels. The image is then converted to a NumPy array and normalized by dividing by 255. The bounding box coordinates and labels are extracted from the DataFrame and converted to NumPy arrays.

Finally, the function returns the images and a dictionary containing the bounding box coordinates and labels, which will be used as the ground truth output for the RCNN model. The labels are encoded using LabelEncoder from scikit-learn to convert them from string labels to integer labels.


```python
import pandas as pd
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load the CSV file containing the bounding boxes and labels
train_boxes_file = 'fracture1/train/_annotations.csv'
val_boxes_file = 'fracture1/valid/_annotations.csv'
train_boxes_df = pd.read_csv(train_boxes_file)
val_boxes_df = pd.read_csv(val_boxes_file)

# Define a function to load the images and their corresponding bounding boxes and labels
def load_data(images_dir, boxes_df):
    images = []
    bboxes = []
    labels = []
    label_encoder = LabelEncoder()
    for idx, row in boxes_df.iterrows():
        image_path = f"{images_dir}/{row['filename']}"
        image = load_img(image_path, target_size=(224, 224))
        image = img_to_array(image) / 255.0
        images.append(image)
        bboxes.append([row['xmin'], row['ymin'], row['xmax'], row['ymax']])
        labels.append(row['class'])
    images = np.array(images)
    bboxes = np.array(bboxes)
    labels = label_encoder.fit_transform(labels)
    return images, {'rcnn_boxes_output': bboxes, 'rcnn_labels_output': labels}

# Load the training and validation images and their corresponding bounding boxes and labels
train_images, train_data = load_data('fracture1/train', train_boxes_df)
val_images, val_data = load_data('fracture1/valid', val_boxes_df)
```

This code defines a combined model that uses a region-based convolutional neural network (RCNN) and a CrackNet model to detect and classify cracks in images.

The RCNN sub-model uses a pre-trained ResNet50 architecture to extract features from the input images, followed by global average pooling and a series of dense layers to generate bounding box coordinates and class probabilities for the input image.

The CrackNet sub-model uses a series of convolutional and pooling layers followed by dense layers and dropout to classify the presence of cracks in the input image.

The combined model uses the outputs of both sub-models, concatenating them to generate a final output that contains the bounding box coordinates, class probabilities, and crack classification.

The model is compiled with three different loss functions, one for each output, and different weights for each loss to balance the importance of each output in the final training.

The optimizer used is Adam, and the model is trained to minimize the combined loss function while maximizing accuracy.

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense,Dropout
# Define the input layer for the combined model
input_layer = Input(shape=(224, 224, 3))

# Define the RCNN sub-model
rcnn_backbone = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
rcnn_output = rcnn_backbone(input_layer)
rcnn_output = tf.keras.layers.GlobalAveragePooling2D()(rcnn_output)
rcnn_output = Dense(1024, activation='relu')(rcnn_output)
rcnn_output = Dense(512, activation='relu')(rcnn_output)
rcnn_boxes_output = Dense(4, activation='linear', name='rcnn_boxes_output')(rcnn_output)
rcnn_labels_output = Dense(num_classes, activation='softmax', name='rcnn_labels_output')(rcnn_output)
# Define the CrackNet sub-model
cracknet_output = Conv2D(16, 3, activation='relu', padding='same')(input_layer)
cracknet_output = MaxPooling2D(2)(cracknet_output)
cracknet_output = Conv2D(32, 3, activation='relu', padding='same')(cracknet_output)
cracknet_output = MaxPooling2D(2)(cracknet_output)
cracknet_output = Conv2D(64, 3, activation='relu', padding='same')(cracknet_output)
cracknet_output = MaxPooling2D(2)(cracknet_output)
cracknet_output = Flatten()(cracknet_output)
cracknet_output = Dense(128, activation='relu')(cracknet_output)
cracknet_output = Dropout(0.5)(cracknet_output)
cracknet_output = Dense(num_classes, activation='softmax', name='cracknet_output')(cracknet_output)
# Define the combined model
combined_output = tf.keras.layers.concatenate([rcnn_boxes_output, rcnn_labels_output, cracknet_output])
combined_model = Model(inputs=input_layer, outputs=combined_output)

# Define the loss function for each output
losses = {
    'rcnn_boxes_output': 'mean_squared_error',
    'rcnn_labels_output': 'sparse_categorical_crossentropy',
    'cracknet_output': 'sparse_categorical_crossentropy'
}

# Define the weight for each output
weights = {
    'rcnn_boxes_output': 1.0,
    'rcnn_labels_output': 1.0,
    'cracknet_output': 0.5
}

# Compile the combined model
combined_model.compile(optimizer='adam', loss=losses, loss_weights=weights, metrics=['accuracy'])

```

This code defines a Region-based Convolutional Neural Network (RCNN) model using the ResNet50 backbone to perform object detection on images. Here's what each part of the code does:

    from tensorflow.keras.applications import ResNet50: Imports the ResNet50 model architecture from Keras applications.
    from tensorflow.keras.layers import Input: Imports the Input layer from Keras layers.
    from tensorflow.keras.models import Model: Imports the Model class from Keras models.
    from tensorflow.keras.layers import Dense: Imports the Dense layer from Keras layers.
    backbone = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3)): Loads the ResNet50 model pre-trained on ImageNet with the last fully-connected layer (top) removed, and specifies the input shape to be (224, 224, 3) which is the shape of the input images.
    input_layer = Input(shape=(224, 224, 3)): Defines the input layer for the RCNN model, which will take images of shape (224, 224, 3) as input.
    backbone_output = backbone(input_layer): Passes the input images through the ResNet50 backbone to obtain the output feature maps.
    rcnn_head = tf.keras.layers.GlobalAveragePooling2D()(backbone_output): Applies global average pooling to the feature maps obtained from the backbone to reduce the spatial dimensions and obtain a fixed-length feature vector.
    rcnn_head = Dense(1024, activation='relu')(rcnn_head): Adds a fully-connected layer with 1024 units and ReLU activation to the RCNN model.
    rcnn_head = Dense(512, activation='relu')(rcnn_head): Adds another fully-connected layer with 512 units and ReLU activation to the RCNN model.
    rcnn_boxes_output = Dense(4, activation='linear', name='rcnn_boxes_output')(rcnn_head): Adds a fully-connected layer with 4 units and linear activation to predict the bounding box coordinates for each object in the input images. The output layer is named rcnn_boxes_output.
    rcnn_labels_output = Dense(num_classes, activation='softmax', name='rcnn_labels_output')(rcnn_head): Adds a fully-connected layer with num_classes units and softmax activation to predict the class probabilities for each object in the input images. The output layer is named rcnn_labels_output.
    rcnn_model = Model(inputs=input_layer, outputs=[rcnn_boxes_output, rcnn_labels_output]): Defines the RCNN model with the input layer as the input and the output layers as the output.
```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense

# Load the pre-trained ResNet50 model as the backbone
backbone = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Define the input layer for the RCNN model
input_layer = Input(shape=(224, 224, 3))

# Pass the input through the ResNet50 backbone
backbone_output = backbone(input_layer)

# Define the RCNN head layers
rcnn_head = tf.keras.layers.GlobalAveragePooling2D()(backbone_output)
rcnn_head = Dense(1024, activation='relu')(rcnn_head)
rcnn_head = Dense(512, activation='relu')(rcnn_head)

# Define the RCNN output layers
rcnn_boxes_output = Dense(4, activation='linear', name='rcnn_boxes_output')(rcnn_head)
rcnn_labels_output = Dense(num_classes, activation='softmax', name='rcnn_labels_output')(rcnn_head)

# Define the RCNN model
rcnn_model = Model(inputs=input_layer, outputs=[rcnn_boxes_output, rcnn_labels_output])
```

The metrics argument is also a dictionary that specifies the metrics to evaluate the model on during training and testing. This code snippet loads a CSV file containing bounding boxes and labels for images, and then defines a function to load the images and their corresponding bounding boxes and labels. The function loads each image from the specified directory, resizes it to (224, 224), converts it to an array and normalizes it by dividing with 255.0. It also extracts the bounding boxes and labels for each image from the DataFrame, encodes the labels using LabelEncoder, and returns the images, bounding boxes, labels, and the LabelEncoder object.

The loaded test images and their corresponding bounding boxes and labels are then passed to the CrackNet_model to make predictions. The predicted labels are then decoded using the inverse_transform method of the LabelEncoder object.

Finally, the predicted images with their corresponding class labels are plotted in a 2x5 grid using matplotlib.

```python
# Load the CSV file containing the bounding boxes and labels
test_boxes_file = 'fracture1/test/_annotations.csv'
test_boxes_df = pd.read_csv(test_boxes_file)

# Define a function to load the images and their corresponding bounding boxes and labels
def load_data(images_dir, boxes_df):
    images = []
    bboxes = []
    labels = []
    label_encoder = LabelEncoder()
    for idx, row in boxes_df.iterrows():
        image_path = f"{images_dir}/{row['filename']}"
        image = load_img(image_path, target_size=(224, 224))
        image = img_to_array(image) / 255.0
        images.append(image)
        bboxes.append([row['xmin'], row['ymin'], row['xmax'], row['ymax']])
        labels.append(row['class'])
    images = np.array(images)
    bboxes = np.array(bboxes)
    labels = label_encoder.fit_transform(labels)
    return images, {'rcnn_boxes_output': bboxes, 'rcnn_labels_output': labels}, label_encoder

# Load the test images and their corresponding bounding boxes and labels
test_images, test_data, label_encoder = load_data('fracture1/test', test_boxes_df)

# Make predictions using CrackNet model
cracknet_preds = CrackNet_model.predict(test_images)

# Decode the predicted labels using the label encoder
predicted_labels = label_encoder.inverse_transform(cracknet_preds.argmax(axis=-1))

# Plot the predicted images with their corresponding class labels
fig, axs = plt.subplots(2, 5, figsize=(15, 8))
for i in range(10):
    ax = axs[i // 5][i % 5]
    ax.imshow(test_images[i])
    ax.axis('off')
    ax.set_title(predicted_labels[i])
plt.show()
```
